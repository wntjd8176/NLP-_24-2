```python
# 크롤링 키워드 및 설정 입력
keword = input("검색할 키워드를 입력해주세요: ")
encText = urllib.parse.quote(keword)

# 크롤링 종료 페이지 설정
end = input("\n크롤링을 끝낼 페이지를 입력해주세요. (기본값:1, 최대값:100): ")
end = int(end) if end else 1
print(f"\n1 ~ {end} 페이지까지 크롤링을 진행합니다.")

# 한 번에 가져올 데이터 수 설정
display = input("\n한 번에 가져올 데이터 개수를 입력해주세요. (기본값:10, 최대값:100): ")
display = int(display) if display else 10
print(f"\n한 번에 가져올 데이터 수: {display}")
```

```python
# 데이터 저장을 위한 리스트 초기화
naver_urls = []
postdate = []
titles = []
```

```python
# 네이버 API 호출
for start in range(1, end + 1):
    url = f"https://openapi.naver.com/v1/search/blog.json?query={encText}&start={start}&display={display}"
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id", client_id)
    request.add_header("X-Naver-Client-Secret", client_secret)

    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if rescode == 200:
        response_body = response.read()
        data = json.loads(response_body.decode('utf-8'))['items']

        for row in data:
            if 'blog.naver' in row['link']:
                naver_urls.append(row['link'])
                postdate.append(row['postdate'])
                titles.append(re.sub('<[^>]*>', '', row['title']))  # HTML 태그 제거
        time.sleep(2)  # 요청 간 지연
    else:
        print(f"Error Code: {rescode}")

        # Chrome 옵션 설정
options = Options()
options.binary_location = "/usr/bin/google-chrome"  # Chrome 실행 파일 경로 (Linux)
# Selenium 초기화
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # 브라우저 없이 실행
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# 블로그 본문 가져오기
contents = []
for url in naver_urls:
    try:
        driver.get(url)
        iframe = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "mainFrame")))
        driver.switch_to.frame(iframe)

        source = driver.page_source
        html = BeautifulSoup(source, "html.parser")
        content = ''.join([c.get_text(separator='\n') for c in html.select("div.se-main-container")])
        contents.append(content)

    except NoSuchWindowException:
        print("브라우저가 닫혀서 드라이버를 재초기화합니다.")
        driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
        driver.get(url)

    except Exception as e:
        print(f"Error occurred: {e}")
        contents.append("error")
        titles.append("error_title")
        postdate.append("unknown_date")

# Selenium 종료
driver.quit()

# 리스트 길이 동기화
max_length = min(len(titles), len(contents), len(postdate))
titles = titles[:max_length]
contents = contents[:max_length]
postdate = postdate[:max_length]

# 데이터 저장
df = pd.DataFrame({'title': titles, 'content': contents, 'date': postdate})
df.to_csv('blog.csv', index=False, encoding='utf-8-sig')

print("크롤링 완료 및 CSV 저장")
```
